# Régression logistique & fonction logistique

```python
import numpy as np
import matplotlib.pyplot as plt
#fonction logistique
def logistic(x):
    # np.exp() calcule l'exponentiel
    return np.exp(x)/(1+np.exp(x))

#Génère 50 valeurs réelles, espacés régulièrement entre -6 et 6
x = np.linspace(-6, 6, 50, dtype=float)

# Transformer chaque nombre en utilisant la fonction logistique
y = logistic(x)

# Tracer les données résultantes
plt.plot(x,y)
plt.ylabel("Probabilité")
plt.show()
```

# Entrainer un modèle de régression logistique

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error

logistic_model = LogisticRegression(solver='liblinear')
logistic_model.fit(admissions[['gpa']], admissions['admit'])
```

# Calcul des probabilités

```python
#predict_proba()
```

```python
pred_probs = logistic_model.predict_proba(admissions[['gpa']])
```

```python
#Probabilités que la ligne appartienne au label '0'
pred_probs[:,0]
```

```python
# Probabilités que la ligne appartienne au label '1'
pred_probs[:,1]
```

```python
import matplotlib.pyplot as plt

plt.scatter(admissions['gpa'], pred_probs[:,1])
```
