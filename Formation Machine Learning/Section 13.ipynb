# Introduction

```python
import pandas as pd
pd.options.mode.chained_assignment = None

data = pd.read_csv('AmesHousing.txt', delimiter="\t")
train = data[0:1460]
test = data[1460:]
df_no_mv = train[train.isnull().sum()[train.isnull().sum() == 0].index]
```

```python
print(train.isnull().sum()[train.isnull().sum() == 0].index)
```

# Caractéristiques de catégorie

```python
temp = df_no_mv.select_dtypes(include=[object]).columns
```

```python
print(temp)
```

```python
train[temp] = train[temp].astype('category')
```

```python
for col in temp:
    print(col,len(train[col].unique()))
```

```python
train['Sale Condition'].value_counts()
```

```python
train['Utilities'].cat.codes.value_counts()
```

```python
train['Utilities'].unique()
```

# Technique pour gérer les caractéristiques de catégorie: colonnes factices

```python
for col in temp:
    col_dummies=pd.get_dummies(train[col])
    train = pd.concat([train, col_dummies], axis=1)
    train.drop(col, axis=1, inplace = True)
    print(col, " supprimé")
    
train.iloc[0:5]
```

```python
train['years_until_remod']=train['Year Remod/Add'] - train['Year Built']
```
