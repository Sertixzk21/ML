```python
airbnb_cols = airbnb.select_dtypes(include=['float']).columns.to_list() #Selectionne les éléments étant en float et organise
#les colonnes en liste
```

```python
for col in airbnb_cols:
    calc_mean = np.mean(airbnb[col])
    print('moyenne de la colonne ' + col + ':', calc_mean)
```
