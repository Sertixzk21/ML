# Fonction pour faire des prédictions

```python
def predict_price(new_listing):
    temp_df=paris_listings.copy()
    #dist = abs(new_listing-temp_df["accommodates"])
    dist = temp_df["accommodates"].apply(lambda x: np.abs(x - new_listing))
    temp_df["distance"] = dist
    temp_df=temp_df.sort_values("distance")
    temp_df["price"]=temp_df["price"].str.replace(",","").str.replace("$","")
    sel=temp_df["price"].astype(float)
 
    predicted_price=np.mean(sel.iloc[0:5])
    return(predicted_price)
    
```

```python
tempo_df=paris_listings.copy()
#disto = abs(1-tempo_df["accommodates"])
disto = tempo_df["accommodates"].apply(lambda x: np.abs(x - 1))
tempo_df["distance"] = disto
tempo_df=tempo_df.sort_values("distance")
tempo_df["price"]=tempo_df["price"].str.replace(",","").str.replace("$","")
selo=tempo_df["price"].astype(float)

predicted_priceo=np.mean(selo.iloc[0:5])
```
